```python
#data structure ; [restaurent_id , 2021 , 2022, 2023,2024]
import numpy as np 
import matplotlib.pyplot as plt
sales_data = np.array([
    [1 , 15000 , 18000 , 22000 , 25000],
    [2 , 12000 , 14000 , 16000 , 19000],
    [3 , 20000 , 23000 , 26000 , 30000],
    [4 , 18000 , 21000 , 24000 , 27000],
    [5 , 16000 , 18500 , 20500 , 23000]
]) 
print("===swigy ===")
print("shape of data = " , sales_data.shape)

print("first 3 restaurent sales:\n",sales_data[:3])
print("\n")
print("all data \n",sales_data[: , 1:])

print("\n")

print("total sales per year: \n" ,np.sum(sales_data,axis=0))#axis = 0 matlab column or axis =1 matlabb row 
year_total = np.sum(sales_data[ : , 1:], axis =0 )
print("total sales for 2021 : \n",year_total)
total_sales = np.sum(year_total)
print("Total sales for 2021 :",total_sales)

min_sales =np.min(sales_data[: , 1:],axis = 1)
print("Minimum Sales  : ", min_sales)

max_sales = np.max(sales_data[: ,1 : ], axis =0)
print("maximum sales for 2021: ",max_sales)
avg = np.mean(sales_data[: , 1 :], axis = 1)
print("average :",avg)
print("\n")
```

```python

cumsum = np.cumsum(sales_data[: , 1: ] ,axis= 1)# cumsum mai ak baad har ak element add hota hai

print("cumsum :",cumsum)
```

```python
plt.figure(figsize=(6,6))
plt.plot(np.mean(cumsum , axis = 0))
plt.title("Average of sales across All restuarrent ")
plt.xlabel("years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
#vector add , multiplication , dot product , Angle between two vector


vector1= np.array([1,2,3,4,5,6])
vector2=np.array([7,8,9,10,11,12])

vector_addition = vector1+vector2
vector_multiply = vector1*vector2
vector_dot_product = np.dot(vector1, vector2)
vector_angle = np.arccos(vector_dot_product/( np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("ector_addition", vector_addition)
print("vector_multiply",vector_multiply)
print("vector_dot_product",vector_dot_product)
print("vector_angle" ,vector_angle)
```
